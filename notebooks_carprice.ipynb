```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df = pd.read_csv('CarPrice.csv')
```

```python
df.columns
```

```python
df['aspiration'] = pd.factorize(df['aspiration'])[0]
df['doornumber'] = pd.factorize(df['doornumber'])[0]
df['drivewheel'] = pd.factorize(df['drivewheel'])[0]
df['cylindernumber'] = pd.factorize(df['cylindernumber'])[0]
df['fuelsystem'] = pd.factorize(df['fuelsystem'])[0]
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df['carprice']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.subplots(figsize=(10, 8))
plt.scatter(df['car_ID'], df['price'])
plt.xlabel('Car ID')
plt.ylabel('Price')
plt.title('Car Prices')
plt.show()
```

```python
plt.subplots(figsize=(10, 8))
plt.hist(df['price'], bins=20)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Car Prices')
plt.show()
```

```python
excluded_cols = ['car_ID', 'symboling']
```

```python
features = df.drop(excluded_cols + ['price'], axis=1)
target = df['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
```

```python
rmse = mean_squared_error(y_test, y_pred, squared=False)
```

```python
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Squared Error: {mse}")
```

```python
print(f"Root Mean Squared Error: {rmse}")
```

```python
print(f"R-squared: {r2}")
```
